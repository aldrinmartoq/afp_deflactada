```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
```

```python
karg_csv = dict(delimiter=';', decimal=',', index_col=0, parse_dates=True)
```

```python
afps = ['CUPRUM', 'HABITAT', 'PLANVITAL', 'PROVIDA', 'MODELO']
fondos = ['A', 'C', 'E']
```

```python
for afp_name in afps:
    filename = 'https://raw.githubusercontent.com/collabmarket/data_afp/master/data/VC-%s.csv'%afp_name
    vc = pd.read_csv(filename, **karg_csv)
    filename = 'https://raw.githubusercontent.com/collabmarket/data_afp/master/data/PAT-%s.csv'%afp_name
    pat = pd.read_csv(filename, **karg_csv)
    # Mayoria Modelo en Fondo B
    if afp_name == 'MODELO':
        fondos = ['A', 'B', 'C', 'E']
    # Ultimos 10 agnos
    N = 365*10
    fig, axes = plt.subplots(nrows=3, figsize=(16,10))
    x = pat[fondos].tail(N)
    titulo = 'Variacion diaria patrimonio AFP %s'%afp_name
    ((x - x.shift()) / x.shift()).plot(ax=axes[0], title=titulo)
    y = vc[fondos].tail(N)
    titulo = 'Variacion diaria valor cuota AFP %s'%afp_name
    y.pct_change().plot(ax=axes[1], sharex=True, title=titulo)
    titulo = 'Valor cuota AFP %s'%afp_name
    y.plot(ax=axes[2], sharex=True, title=titulo)
    plt.savefig('cambios/cambios-%s.png'%afp_name)
    plt.show()
```

```python
N = 365*4
fig, axes = plt.subplots(nrows=3, figsize=(16,10))
x = pat[fondos].tail(N)
titulo = 'Variacion diaria patrimonio AFP %s'%afp_name
((x - x.shift()) / x.shift()).plot(ax=axes[0], title=titulo)
y = vc[fondos].tail(N)
titulo = 'Variacion diaria valor cuota AFP %s'%afp_name
y.pct_change().plot(ax=axes[1], sharex=True, title=titulo)
titulo = 'Valor cuota AFP %s'%afp_name
y.plot(ax=axes[2], sharex=True, title=titulo)
plt.show()
```
